## Mid Term 1 data Reader </br>
Notebook that reads the netcdf file giving the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
```

```python
# Start by reading in the data the sample space N and the state space M
ncfile ='test2.nc' 
ds = nc.Dataset(ncfile)
print(ds.variables.keys()) # get all variable names

xnc = ds.variables['value']
space = ds.variables['space']
time = ds.variables['time']
x = np.ma.getdata(xnc, subok=False)

print(space)
print(time)
print(x)
print('type x ',type(x))

plt.figure()
plt.contourf(x)
plt.colorbar
plt.ylabel('Time')
plt.xlabel('space')
```
